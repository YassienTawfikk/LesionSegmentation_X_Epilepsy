```python
import shutil
import os
import pandas
```

```python
def clear_kaggle_working():
    path = "/kaggle/working"
    for item in os.listdir(path):# Show the input Data Structure
        item_path = os.path.join(path, item)
        try:
            if os.path.isfile(item_path) or os.path.islink(item_path):
                os.unlink(item_path)
            elif os.path.isdir(item_path):
                shutil.rmtree(item_path)
        except Exception as e:
            print(f"Failed to delete {item_path}: {e}")


def print_folder_structure(startpath, indent=""):
    """
    Recursively prints the directory structure from the given startpath.

    Args:
        startpath (str): The path to the directory to start exploring.
        indent (str): Used for indentation to visualize the hierarchy.
    """
    if not os.path.exists(startpath):
        print(f"Error: Path '{startpath}' does not exist.")
        return

    if not os.path.isdir(startpath):
        print(f"Error: Path '{startpath}' is not a directory.")
        return

    # Get all entries (files and directories) in the current path
    entries = os.listdir(startpath)
    entries.sort() # Sort alphabetically for consistent output

    for i, entry in enumerate(entries):
        path = os.path.join(startpath, entry)
        is_last_entry = (i == len(entries) - 1)

        # Determine the prefix for the current entry (tree-like structure)
        prefix = "â””â”€â”€ " if is_last_entry else "â”œâ”€â”€ "
        print(f"{indent}{prefix}{entry}")

        # If it's a directory, recurse into it
        if os.path.isdir(path):
            next_indent = indent + ("    " if is_last_entry else "â”‚   ")
            print_folder_structure(path, next_indent)
```

## Predict Subject number 3 Using GPU

```python
import os
from SynthSeg.predict import predict

# --- Custom Paths and Settings ---

# The path to your input image
path_images = "/kaggle/input/fcd-subject-centric/FCD-Subject-Centric Dataset/sub-00003/t1.nii"

# The home directory for the SynthSeg repository
synthseg_home = "/kaggle/working/SynthSeg-Modern"

# The desired output directory (will be created if it doesn't exist)
output_dir = "/kaggle/working/synthseg_output"

# Set this to TRUE to enable GPU execution
use_gpu = True

# --- Path Construction and Setup ---

# Create the output directory if it doesn't exist
os.makedirs(output_dir, exist_ok=True)

# Path to the trained model (adjusted based on your installation path)
path_model = os.path.join(synthseg_home, "models", "synthseg_1.0.h5")

# Paths for the output files
base_name = os.path.basename(path_images).replace(".nii.gz", "").replace(".nii", "")
path_segm = os.path.join(output_dir, f"{base_name}_seg.nii.gz")
path_posteriors = os.path.join(output_dir, f"{base_name}_post.nii.gz")
path_resampled = os.path.join(output_dir, f"{base_name}_resampled.nii.gz")
path_vol = os.path.join(output_dir, "volumes.csv")

# Paths to auxiliary files (relative to synthseg_home)
data_dir = os.path.join(synthseg_home, "data", "labels_classes_priors")
path_segmentation_labels = os.path.join(data_dir, "synthseg_segmentation_labels.npy")
path_segmentation_names = os.path.join(data_dir, "synthseg_segmentation_names.npy")
topology_classes = os.path.join(data_dir, "synthseg_topological_classes.npy")

# --- Preprocessing Parameters ---
cropping = 192
target_res = 1.0

# --- Postprocessing Parameters ---
flip = True
n_neutral_labels = 18
sigma_smoothing = 0.5
keep_biggest_component = True

# --- Network Architecture Parameters (must match training) ---
n_levels = 5
nb_conv_per_level = 2
conv_size = 3
unet_feat_count = 24
activation = 'elu'
feat_multiplier = 2

# --- Evaluation Parameters (Set to None as ground truth path is not provided) ---
gt_folder = None
compute_distances = False

# --- Core Logic: Configure CPU/GPU and Run Prediction ---
print("--- Starting SynthSeg Prediction (4-prediction.py) ---")
print(f"Input Image: {path_images}")
print(f"Output Segmentation: {path_segm}")
print(f"Model Path: {path_model}")

if not use_gpu:
    # This block is now skipped since use_gpu is True.
    # We will let TensorFlow automatically find and use the GPU.
    os.environ['CUDA_VISIBLE_DEVICES'] = '-1'
    print("\nâš ï¸ **FORCING CPU EXECUTION (CUDA_VISIBLE_DEVICES=-1)**\n")
else:
    # Standard practice to ensure TensorFlow uses an available GPU
    # If multiple GPUs exist, it will use the default one (usually device 0).
    os.environ['CUDA_VISIBLE_DEVICES'] = '0'
    print("\nâœ… **ENABLING GPU EXECUTION (CUDA_VISIBLE_DEVICES=0)**\n")

# All right, we're ready to make predictions !!
predict(path_images,
        path_segm,
        path_model,
        path_segmentation_labels,
        n_neutral_labels=n_neutral_labels,
        path_posteriors=path_posteriors,
        path_resampled=path_resampled,
        path_volumes=path_vol,
        names_segmentation=path_segmentation_names,
        cropping=cropping,
        target_res=target_res,
        flip=flip,
        topology_classes=topology_classes,
        sigma_smoothing=sigma_smoothing,
        keep_biggest_component=keep_biggest_component,
        n_levels=n_levels,
        nb_conv_per_level=nb_conv_per_level,
        conv_size=conv_size,
        unet_feat_count=unet_feat_count,
        feat_multiplier=feat_multiplier,
        activation=activation,
        gt_folder=gt_folder,
        compute_distances=compute_distances)

print("\nğŸ‰ GPU prediction script executed. Check the output directory for results.")
```
