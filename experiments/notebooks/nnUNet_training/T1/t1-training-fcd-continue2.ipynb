```python
import os

# Define base paths
base_input = "/kaggle/input/t1-training-fcd-dataset"
base_working = "/kaggle/working"

# Define nnU-Net paths inside working directory
nnunet_raw = os.path.join(base_working, "nnUNet_raw")
nnunet_preprocessed = os.path.join(base_working, "nnUNet_preprocessed")
nnunet_results = os.path.join(base_working, "nnUNet_results")

# Create empty raw folder (even if not used)
os.makedirs(nnunet_raw, exist_ok=True)
os.makedirs(nnunet_preprocessed, exist_ok=True)
os.makedirs(nnunet_results, exist_ok=True)

# Set environment variables
os.environ["nnUNet_raw"] = nnunet_raw
os.environ["nnUNet_preprocessed"] = nnunet_preprocessed
os.environ["nnUNet_results"] = nnunet_results

# For torch on Kaggle GPU (P100)
os.environ["TORCH_COMPILE_DISABLE"] = "1"

print("✅ Environment variables set")
```
