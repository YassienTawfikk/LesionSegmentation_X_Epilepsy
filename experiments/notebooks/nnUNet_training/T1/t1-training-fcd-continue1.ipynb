```python
import os

os.environ["nnUNet_raw"] = "/kaggle/working/nnUNet_raw_data"
os.environ["nnUNet_preprocessed"] = "/kaggle/working/nnUNet_preprocessed"
os.environ["nnUNet_results"] = "/kaggle/working/nnUNet_results"

print("Environment variables set ✅")
```

```python
import shutil

src_base = "/kaggle/input/t1-training-fcd-v1"
dst_base = "/kaggle/working"

# 1. Copy nnUNet_results
shutil.copytree(f"{src_base}/nnUNet_results", f"{dst_base}/nnUNet_results", dirs_exist_ok=True)

# 2. Copy nnUNet_preprocessed (inside preprocessed-dataset-t1)
shutil.copytree(f"{src_base}/preprocessed-dataset-t1/nnUNet_preprocessed", 
                f"{dst_base}/nnUNet_preprocessed", dirs_exist_ok=True)

print("✅ Copied nnUNet_results and nnUNet_preprocessed (raw data skipped)")
```

```python
import os

# Disable torch.compile for P100 compatibility
os.environ["NNUNET_USE_TORCH_COMPILE"] = "0"
os.environ["TORCH_COMPILE_DISABLE"] = "1"
os.environ["CUDA_VISIBLE_DEVICES"] = "0"

import torch._dynamo
torch._dynamo.config.suppress_errors = True

print("Torch Compile disabled. Ready to train nnU-Net safely on Tesla P100 ✅")
```
